###### 18. Write a Pandas program to get sample 75% of the diamonds DataFrame's rows without replacement and store the remaining 25% of the rows in another DataFrame.

```python
print("\nsample 75% of diamonds Dataframe row without replacement: ")
result=data.sample(frac=0.75,random_state=99)
print(result)
print("\nRemaining 25% of the rows: ")
print(data.loc[~data.index.isin(result.index), :])
```
